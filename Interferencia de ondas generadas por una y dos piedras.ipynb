Onda generada por una piedra

```python
x=linspace(0,100,1)
y=linspace(0,100,1)
xlim(0.,500)
ylim(0.,500)
c=1
b=5
k=(2*3.1416)/b
x1=0
y1=0
nx=500
dimx=100
dx=dimx/nx
salida=open("ondas_est.dat","w")

for i in range(0,501) :
    for j in range(0,501):
        r1=sqrt((x1-i*dx)**2+(y1-j*dx)**2)
        C=c*sin(k*r1)
        salida.write("%0.2f "% (c*sin(k*r1)))
    salida.write("\n")
salida.close()
print(C)
data = loadtxt("ondas_est.dat", float)
print(data.shape)
imshow(data)
show()
```

Onda generada por dos piedras

```python
x=linspace(0,500,1)
y=linspace(0,500,1)
c=1
b=5
k=(2*3.1416)/b
x1=50
y1=25
x2=50
y2=75
nx=500
dimx=100
dx=dimx/nx
salida=open("ondas_est.dat","w")

for i in range(0,501) :
    for j in range(0,501):
        r1=sqrt((x1-i*dx)**2+(y1-j*dx)**2)
        r2=sqrt((x2-i*dx)**2+(y2-j*dx)**2)
        C1=c*sin(k*r1)
        C2=c*sin(k*r2)
        C=C1+C2
        salida.write("%0.2f "% C)
    salida.write("\n")
salida.close()
print(C)
data = loadtxt("ondas_est.dat", float)
imshow(data, origin="lower", extent=[0, 100, 0, 100])
print(data.shape)
show()
```

```python
salida2=open("ondas_estanque.dat","w")
s=400
d=100
x=linspace(0,d,s)
y=linspace(0,d,s)
X,Y=np.meshgrid(x,y)
c1=1 #amplitud
c2=2 #aplitud particula 2
w1=2
w2=5#frecuencia
b1=5 #longitud de onda
b2=3
k1=(2*3.1416)/b1
k2=(2*3.1416)/b2
x1=25 #psicion inicial
y1=25 #posicion inicial y
x2=75 #posisicion inicial paticula 2
y2=55 #posisicion inicial y
t=20 #tiempo
info=np.zeros((s,s,t),dtype=float)
dimx=100
nx=500
dx=dimx/nx
for t in range(0,t):
    r1=sqrt((X-x1)**2 +(Y-y1)**2)
    r2=sqrt((X-x2)**2 +(Y-y2)**2)
    C=c1*sin(k1*r1+w1*t)+c2*sin(k2*r2+w2*t)
    plt.gca().cla()
    info[:,:,t]=C
    imshow(C,origin="lower", extent=[0, d, 0, d])
    hot()
    disp.clear_output(wait=True)
    disp.display(plt.gcf())
    
    
```
